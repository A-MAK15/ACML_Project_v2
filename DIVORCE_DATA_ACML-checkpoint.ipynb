```python
X= df.drop(["GroundsForDivorce"], axis = 1)
y = df[["GroundsForDivorce"]]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state = 100) #80% training, 20% testing
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
rfc = RandomForestClassifier(criterion="entropy", max_depth=8, min_samples_split=10, random_state=42, verbose=True)
dt_model = DecisionTreeClassifier(max_depth=6)
xgb_model = XGBClassifier(objective="multi:softprob", num_class=8, n_estimators=300, max_depth=None, learning_rate=0.01,
    subsample=0.8, colsample_bytree=0.8, random_state=100)
```

```python
rfc.fit(X_train, y_train)
```

```python
dt_model.fit(X_train, y_train)
```

```python
encoder = LabelEncoder()
y_train_encoded = encoder.fit_transform(y_train)
y_test_encoded = encoder.transform(y_test)
```

```python
xgb_model.fit(X_train, y_train_encoded)
```

```python
y_pred_dt = dt_model.predict(X_test)
y_pred_rfc = rfc.predict(X_test)
y_pred_xgb = xgb_model.predict(X_test)
```

```python
#Decision Tree
accuracy = accuracy_score(y_test, y_pred_dt) 
print("Accuracy : {:.2f}%".format(accuracy*100))
```

```python
#Random Forest Model Evaluation
print(confusion_matrix(y_test, y_pred_rfc))
print()
print("Accuracy Score")
print(accuracy_score(y_test, y_pred_rfc))
print()
print("Classification report")
print(classification_report(y_test, y_pred_rfc))
print("Confusion matrix")
sns.heatmap(confusion_matrix(y_test, y_pred_rfc), annot=True)
```

```python
print(f"Accuracy Score : {accuracy_score(y_test_encoded, y_pred_xgb)}")
```
