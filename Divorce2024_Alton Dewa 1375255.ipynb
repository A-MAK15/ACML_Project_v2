### Perform one-hot encoding on the identified categorical columns in the training set.

### Apply this same transformation to  validation and test sets to maintain consistency.

```python
# Identify categorical columns to be one-hot encoded
# These are columns that represent categories, even if they are integers.
categorical_cols = [
    'ProvinceCode', 'MaleRace', 'FemaleRace', 'CoupleRace',
    'MaleOccupation', 'FemaleOccupation', # MaleOccupation and FemaleOccupation contain NaNs which get_dummies can handle as a category
    'PlaintiffSex', 'MaleStatus', 'FemaleStatus', 'SolemnisationCode',
    'MarriagePropertySystem', 'MarriageMonth', 'DivorceMonth', 'MarriageType',
    'PolygamousMarriage', 'GroundsForDivorce_Label'
]

# Convert identified integer columns to 'category' dtype to ensure get_dummies treats them as such
# This is crucial for columns that are integers but represent categories.
for col in categorical_cols:
    if col in train_df.columns:
        train_df[col] = train_df[col].astype('category')
    if col in validation_df.columns:
        validation_df[col] = validation_df[col].astype('category')
    if col in test_df.columns:
        test_df[col] = test_df[col].astype('category')

# Perform one-hot encoding on the training set
# Using dummy_na=True for MaleOccupation and FemaleOccupation to treat NaN as a separate category
train_df_encoded = pd.get_dummies(train_df, columns=categorical_cols, dummy_na=False)

# Display the shape of the new DataFrame and the first few rows
print(f"Shape of training set after one-hot encoding: {train_df_encoded.shape}")
display(train_df_encoded.head())

# Note: It's important to apply the same one-hot encoding transformation to the validation and test sets.
# To ensure consistent columns across all splits, we should fit the OneHotEncoder on the training data
# and then transform all sets, or use pd.get_dummies with careful handling of columns.
# For this demonstration, we'll show the process for the training set. For a robust ML pipeline,
# ensure all splits have the exact same columns after encoding.

# Let's perform the same for validation and test sets, ensuring consistent columns
# For simplicity, we will re-run get_dummies on all, then align columns.
validation_df_encoded = pd.get_dummies(validation_df, columns=categorical_cols, dummy_na=False)
test_df_encoded = pd.get_dummies(test_df, columns=categorical_cols, dummy_na=False)

# Align columns - this step is crucial to ensure all dataframes have the same columns
# after one-hot encoding, especially if some categories are not present in all splits.
common_cols = list(set(train_df_encoded.columns) & set(validation_df_encoded.columns) & set(test_df_encoded.columns))

train_df_final = train_df_encoded[common_cols]
validation_df_final = validation_df_encoded[common_cols]
test_df_final = test_df_encoded[common_cols]

# Display the shapes of the final aligned DataFrames
print(f"\nShape of final aligned training set: {train_df_final.shape}")
print(f"Shape of final aligned validation set: {validation_df_final.shape}")
print(f"Shape of final aligned test set: {test_df_final.shape}")
```

### Addressing Class Imbalance with SMOTE (Synthetic Minority Over-sampling Technique)

Despite using `class_weight='balanced'` in the RandomForestClassifier, the model still shows very poor performance on the most imbalanced (minority) classes. This suggests that simply adjusting weights might not be sufficient when the minority classes have extremely few samples. To further address this, we will apply **SMOTE (Synthetic Minority Over-sampling Technique)** to the training data.

SMOTE works by creating synthetic samples from the minority class. It does this by taking sampl.es from the minority class, and for each sample, finding its k-nearest neighbors. It then creates new synthetic samples by taking the difference between the feature vector of the current sample and its neighbors, multiplying by a random number between 0 and 1, and adding it to the current feature vector. This effectively 'smears' the minority class into the feature space, making it more visible to the classifier.

We will apply SMOTE only to the training data to avoid data leakage from the validation and test sets
