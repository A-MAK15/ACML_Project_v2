```python
plt.figure(figsize=(8,5))
plt.hist(df['HusbandAge'], bins=50)
plt.title("Distribution of Husband Age (Raw)")
plt.xlabel("Husband Age")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(8,5))
plt.hist(df['WifeAge'], bins=50)
plt.title("Distribution of Wife Age (Raw)")
plt.xlabel("Wife Age")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.boxplot(df['DurationOfMarriage'], vert=False)
plt.title("Boxplot of Duration of Marriage")
plt.xlabel("Duration of Marriage (years)")
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.scatter(df['HusbandAge'], df['DurationOfMarriage'], alpha=0.4)
plt.title("Husband Age vs Duration of Marriage")
plt.xlabel("Husband Age")
plt.ylabel("Duration of Marriage")
plt.show()

plt.figure(figsize=(7,5))
plt.scatter(df['WifeAge'], df['DurationOfMarriage'], alpha=0.4)
plt.title("Wife Age vs Duration of Marriage")
plt.xlabel("Wife Age")
plt.ylabel("Duration of Marriage")
plt.show()
```

```python
province_mean = df.groupby("ProvinceCode")["DurationOfMarriage"].mean().sort_values()

plt.figure(figsize=(10,5))
plt.bar(province_mean.index.astype(str), province_mean.values)
plt.title("Average Duration of Marriage by Province")
plt.xlabel("Province Code")
plt.ylabel("Mean Duration (years)")
plt.show()
```

```python
#Replace outliers with NaN
df["HusbandAge"] = df["HusbandAge"].replace(999, np.nan)
df["WifeAge"] = df["WifeAge"].replace(999, np.nan)

df["DurationOfMarriage"] = df["DurationOfMarriage"].replace(99, np.nan)
```

```python
#Convert NaN to integers
df["HusbandAge_missing"] = df["HusbandAge"].isna().astype(int)
df["WifeAge_missing"] = df["WifeAge"].isna().astype(int)
df["Duration_missing"] = df["DurationOfMarriage"].isna().astype(int)

#Impute by filling in missing values with median
df["HusbandAge"] = df["HusbandAge"].fillna(df["HusbandAge"].median())
df["WifeAge"] = df["WifeAge"].fillna(df["WifeAge"].median())
```
